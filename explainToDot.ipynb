```python
maxRows = 0
def trackMaxRows(r):
    global maxRows
    
    thisCount = float(r)
    if thisCount > maxRows:
        maxRows = thisCount

def setOpAttribute(thisOp, data):
    chunks = data.split('(')
    chunks = [c.replace(')', '').rstrip().lstrip() for c in chunks[1:]]
    if len(chunks) == 2:
        # The case for an extra chunk containing 
        # slice and segment metadata
        sliceSegmentTokens = chunks[0].split(';')
        tokens = chunks[1].split() + sliceSegmentTokens
    else:
        tokens = chunks[0].split()
        
    for t in tokens:
        if '=' in t:
            # Typical case
            key, val = t.split('=')
        elif ':' in t:
            # Number of segments
            key, val = t.split(':')
        else:
            # Slice identifier
            key = t[:-1]
            val = t[-1]
            
        # Clean whitespace
        key = key.lstrip().rstrip()
        val = val.lstrip().rstrip()
        
        if key == 'width':
            key = 'byteWidth'
        
        #graph.get_node(thisOp).attr[key] = val
        graph.node[thisOp][key] = val
        
        if key == 'rows':
            trackMaxRows(val)
            graph.node[thisOp]['weight'] = val
            
        print thisOp, "|", key, "|", val
            

def setSubAttribute(thisOp, data):
    try:
        name, val = data.split(':', 1)
    except ValueError:
        print "|", data, "|"
        return
    
    key = name.lstrip().rstrip()
    val = val.lstrip().rstrip()
    
    if key == 'width':
        key = 'byteWidth'

    if "Rows" in key:
        # Average rows is the 2nd token
        # when split by spaces
        numRows = 0
        if "Gather Motion" in thisOp:
            numRows = val.split()[0]
        else:
            numRows = val.split()[1]
        
        trackMaxRows(numRows)
        graph.node[thisOp]['weight'] = numRows
        
    #graph.get_node(thisOp).attr[key] = val
    graph.node[thisOp][key] = val  
    
    print thisOp, "|", key, "|", val
        
        
#graph = pgv.AGraph(strict=False,directed=True)
graph = nx.Graph(name='Query Plan',
                 size="7.75,10.25",
                 ordering='in',
                 rankdir='RL')

levelList = list()
level = 0
levelTrack = dict()
# Track parent Op
parentOp = '{}-{}'.format(0, ops[1].split('(')[0].lstrip().rstrip())
print "Root:", parentOp
thisOp = parentOp

# Root node
graph.add_node(parentOp, label=thisOp.split('-')[1], rank='min')
setOpAttribute(thisOp, ops[1])

for n, line in enumerate(ops[2:]):
    curToken = ' '
    i = 0
    # Count spaces before the first character
    # to determine Op depth level
    while curToken == ' ' and i < len(line):
        curToken = line[i]
        i = i + 1
    
    if curToken == '-':
        # If this level has been seen, look it up
        if i in levelTrack:
            level = levelTrack[i]['level'] #+ 0.1
            
            # If this is deeper than the last
            # seen level, parent is previous Op
            if level > levelList[-1]:
                parentOp = thisOp
            else:
                parentOp = levelTrack[i]['parent']
        # Reached a new level, increment and track it
        else:
            level = level + 1
            parentOp = thisOp
        
        thisOp = '{}-{}'.format(level, line.split('(')[0].rstrip().split('>')[1].lstrip())
        #print level, parentOp, thisOp
        levelTrack[i] = {'level': level, 'parent': parentOp}
        levelList.append(level)
        
        # TODO
        # Need to de-couple levels and node names (use Op for node name)
        graph.add_node(thisOp, label=thisOp.split('-')[1])
        setOpAttribute(thisOp, line)
        numRows = graph.node[thisOp]['rows']
        graph.add_edge(thisOp, parentOp, label=numRows, weight=numRows)
    # This is metadata for the previous Op
    else:
        levelList.append(level)
        
        setSubAttribute(thisOp, line)
        
print "Max rows:", maxRows
graph.graph ['maxRows'] = maxRows
```

```python
%%javascript
// We load the d3.js library from the Web.
require.config({paths: {d3: "http://d3js.org/d3.v3.min"}});
require(["d3"], function(d3) {
    // The code in this block is executed when the 
    // d3.js library has been loaded.
    
    // First, we specify the size of the canvas containing
    // the visualization (size of the <div> element).
    var width = 1000,
        height = 600;

    // We create a color scale.
    var color = d3.scale.category10();

    // We create a force-directed dynamic graph layout.
    var force = d3.layout.force()
        .charge(-200)
        .linkDistance(60)
        .size([width, height]);

    // In the <div> element, we create a <svg> graphic
    // that will contain our interactive visualization.
    var svg = d3.select("#d3-example").select("svg")
    if (svg.empty()) {
        svg = d3.select("#d3-example").append("svg")
                    .attr("width", width)
                    .attr("height", height);
    }
        
    // We load the JSON file.
    d3.json("explainGraph.json", function(error, graph) {
        // In this block, the file has been loaded
        // and the 'graph' object contains our graph.
        
        // We load the nodes and links in the force-directed
        // graph.
        force.nodes(graph.nodes)
            .links(graph.links)
            .start();

        // We create a <line> SVG element for each link
        // in the graph.
        var link = svg.selectAll(".link")
            .data(graph.links)
            .enter().append("line")
            .attr("class", "link");

        // We create a <circle> SVG element for each node
        // in the graph, and we specify a few attributes.
        var node = svg.selectAll(".node")
            .data(graph.nodes)
            .enter().append("circle")
            .attr("class", "node")
            .attr("r", 12)  // radius
            .style("fill", function(d) {
                // The node color depends on the club.
                return 'red'; //color('white'); //d.club); 
            })
            .call(force.drag);

        // The name of each node is the node number.
        node.append("text")
            .attr("dy", ".35em")
            .text(function(d) { return d.label; });

        // We bind the positions of the SVG elements
        // to the positions of the dynamic force-directed graph,
        // at each time step.
        force.on("tick", function() {
            link.attr("x1", function(d) { return d.source.x; })
                .attr("y1", function(d) { return d.source.y; })
                .attr("x2", function(d) { return d.target.x; })
                .attr("y2", function(d) { return d.target.y; });

            node.attr("cx", function(d) { return d.x; })
                .attr("cy", function(d) { return d.y; });
        });
    });
});
```
