```python
# Method and globals to track [max] row counts
opRowDict = dict()
maxRows = 0
def trackMaxRows(thisOp, r):
    global maxRows
    
    thisCount = float(r)
    opRowDict[thisOp] = thisCount
    
    if thisCount > maxRows:
        maxRows = thisCount
        
def makeTableLabel(thisOp):
    # Open with a table tag
    labelString = '<<table border="1" cellspacing="0">\n'
    
    # Get the dictionary of node attribute
    attrs = graph.get_node(thisOp).attr
    
    # Make sure Label is first
    labelString += '\t<tr><td border="2" colspan="2"><b>{}</b></td></tr>\n'.format(attrs['label'])
    
    # Build a table using all key-val pairs of set attributes
    for k, v in attrs.iteritems():
        # Ignore the label key, it was set as first row
        if k == 'label':
            continue
        labelString += '\t<tr><td>{}</td><td>{}</td></tr>\n'.format(k, v.replace('<', '&lt;').replace('>', '&gt;'))
        
        # How to color things?
        # bgcolor="red"
        
    # Close the table
    labelString += '</table>>'
    
    # Set shape to none and apply the HTML
    # for GraphViz to render correctly
    attrs['shape'] = 'none'
    attrs['label'] = labelString
    
    #print labelString
        
# Method for setting attributes using metadata
# contained in the same line as the op name
def setOpAttribute(thisOp, data):
    # Set the node label using the operation name
    graph.get_node(thisOp).attr['label'] = thisOp.split('-')[1]
    
    # Chunk data with l-paren as delimiter
    chunks = data.split('(')
    
    # Remove r-parens and strip leading/trailing whitespace, skipping
    # the first chunk since it is the Op label, which is already stored
    chunks = [c.replace(')', '').rstrip().lstrip() for c in chunks[1:]]
    if len(chunks) == 2:
        # The case for an extra chunk containing 
        # slice and segment metadata
        sliceSegmentTokens = chunks[0].split(';')
        tokens = chunks[1].split() + sliceSegmentTokens
    else:
        tokens = chunks[0].split()
        
    # Set Segment Motion In/Out metadata
    if 'Redistribute' in thisOp:
        # Segments are specified as the last space-delimited 
        # token in the op name, formatted as in:out [segments]
        segsIn, segsOut = thisOp.split()[-1].split(':')
        tokens.append('segsIn={}'.format(segsIn))
        tokens.append('segsOut={}'.format(segsOut))
        
    for t in tokens:
        if '=' in t:
            # Typical case
            key, val = t.split('=')
        elif ':' in t:
            # Case to get number of segments
            key, val = t.split(':')
        else:
            # Case to get the slice identifier
            key = t[:-1]
            val = t[-1]
            
        # Clean whitespace
        key = key.lstrip().rstrip()
        val = val.lstrip().rstrip()
        
        # Catch this key since width means
        # something different to GraphViz
        if key == 'width':
            key = 'byteWidth'
        
        # PyGraphViz API
        graph.get_node(thisOp).attr[key] = val
        
        # NetworkX API
        #graph.node[thisOp][key] = val
        
    # What is the difference between rows specified
    # with the operationlabel  versus in the op metadata?
    #if 'rows' in graph.get_node(thisOp).attr:

        #trackMaxRows(thisOp, val)

        # PyGraphViz API
        #graph.get_node(thisOp).attr['weight'] = float(val)

        # NetworkX API
        #graph.node[thisOp]['weight'] = val
            
        #print thisOp, "|", key, "|", val
            
# Method for setting attributes using metadata
# listed underneath a particular op, parent op
# is needed so the edge can be labeled with row data
def setSubAttribute(thisOp, parentOp, data):
    try:
        name, val = data.split(':', 1)
    except ValueError:
        print "Ignoring line :", "|", data.lstrip(), "|"
        return
    
    key = name.lstrip().rstrip()
    val = val.lstrip().rstrip()
    
    if key == 'width':
        key = 'byteWidth'

    if "Rows" in key:
        numRows = 0
        # Rows is the 1st token
        # when split by spaces
        if "Gather Motion" in thisOp:
            numRows = val.split()[0]
        # Average rows is the 2nd token
        # when split by spaces
        else:
            numRows = val.split()[1]
        
            # In the case of redistributing, rows are multiplied
            # by the number of segments receiving from 
            if "Redistribute" in thisOp:
                multiplier = int(graph.get_node(thisOp).attr['segsIn'])
                print numRows, multiplier
                numRows = float(numRows)*multiplier
                print numRows

        
        trackMaxRows(thisOp, numRows)
        
        # This Try-Except block catches the case where
        # no edges exist yet (root node)
        try:
            graph.get_edge(thisOp, parentOp).attr['label'] = float(numRows)
        except KeyError:
            pass
        
        # PyGraphViz API
        graph.get_node(thisOp).attr['rows'] = float(numRows)
        
        # NetworkX API
        #graph.node[thisOp]['weight'] = numRows
        
    # PyGraphViz API
    graph.get_node(thisOp).attr[key] = val
    
    # NetworkX API
    #graph.node[thisOp][key] = val  
    
    #print thisOp, "|", key, "|", val
        
        
# PyGraphViz API
graph = pgv.AGraph(strict=False,
                   directed=True)
                   #rankdir='LR')
    
# NetworkX API
#graph = nx.Graph(name='Query Plan',
#                 size="7.75,10.25",
#                 ordering='in',
#                 rankdir='RL')

# Helper globals
levelList = list()
level = 0
levelTrack = dict()
dotRankMembers = dict()

# Track the parent Op
parentOp = '{}-{}'.format(0, ops[1].split('(')[0].lstrip().rstrip())
#print "Root:", parentOp
thisOp = parentOp

# Add the root node
graph.add_node(thisOp, label=thisOp.split('-')[1], rank='min')
setOpAttribute(thisOp, ops[1])

for n, line in enumerate(ops[2:]):
    curToken = ' '
    i = 0
    # Count spaces before the first character
    # to determine the op depth level
    while curToken == ' ' and i < len(line):
        curToken = line[i]
        i = i + 1
    
    # Token marks a new op boundary (e.g.   -> Hash Join)
    if curToken == '-':
        # Set the label for the previous Op (still thisOp at this point)
        makeTableLabel(thisOp)
        
        # If this level has been seen before, look it up
        if i in levelTrack:
            level = levelTrack[i]['level']
            
            # If level is deeper than the last
            # seen level, parent is previous Op
            if level > levelList[-1]:
                parentOp = thisOp
            
            # Otherwise the parent is shared with
            # a previously seen op
            else:
                parentOp = levelTrack[i]['parent']
        # Reached a new level, increment and track it
        else:
            level = level + 1
            
            # This may not be needed as long as the
            # graph remains directed -- for display purposes
            dotRankMembers[level] = list()
            
            parentOp = thisOp
        
        # Set a unique op name using the current level and the operation name
        operationName = line.split('(')[0].rstrip().split('>')[1].lstrip()
        thisOp = '{}-{}'.format(level, operationName)
        
        # For display purposes, track all ops
        # that should be displayed at the same rank level
        # (i.e. same tab depth in raw output)
        dotRankMembers[level].append(thisOp)
        
        #print level, parentOp, thisOp
        levelTrack[i] = {'level': level, 'parent': parentOp}
        levelList.append(level)
        
        # NetworkX API
        #graph.add_node(thisOp, label=thisOp.split('-')[1])
        #setOpAttribute(thisOp, line)
        #numRows = graph.node[thisOp]['rows']
        #graph.add_edge(thisOp, parentOp) #, label=numRows, weight=numRows)
        
        # PyGraphViz API
        graph.add_edge(thisOp, parentOp) 
        setOpAttribute(thisOp, line)
    # This is metadata for the previous Op
    else:
        levelList.append(level)
        
        setSubAttribute(thisOp, parentOp, line)
        
# Set the label for the final Op
makeTableLabel(thisOp)
        
print "Max rows:", maxRows
#graph.graph ['maxRows'] = maxRows
```

```python
equiLevelRanks = list()
for r in dotRankMembers.keys():
    #print dotRankMembers[r]
    equiLevelRanks.append(dotRankMembers[r])

def graphvizRankedLayout(G, prog = "neato", root = None, sameRank = [], args = ""):
    try:
        import pygraphviz
    except ImportError:
        raise ImportError('requires pygraphviz ',
                          'http://networkx.lanl.gov/pygraphviz ',
                          '(not available for Python3)')
    if root is not None:
        args+="-Groot='%s'"%root
        
    A = G
    for sameNodeHeight in sameRank:
        if type(sameNodeHeight) == str:
            print("node \"%s\" has no peers in its rank group" %sameNodeHeight)
        A.add_subgraph(sameNodeHeight, rank="same")
    print args
    A.layout(prog=prog, args=args)
    
    node_pos = {}
    for n in A:
        node = pygraphviz.Node(A,n)
        try:
            xx,yy = node.attr["pos"].split(',')
            node_pos[n] = (float(xx),float(yy))
        except:
            print("no position for node",n)
            node_pos[n] = (0.0,0.0)
            
    return node_pos, A

rankedPos, newGraph = graphvizRankedLayout(graph, 
                                    prog='dot', 
                                    root=equiLevelRanks[0][0], 
                                    sameRank=equiLevelRanks[1:])
```

```python
%%javascript
// We load the d3.js library from the Web.
require.config({paths: {d3: "http://d3js.org/d3.v3.min"}});
require(["d3"], function(d3) {
    // The code in this block is executed when the 
    // d3.js library has been loaded.
    
    // First, we specify the size of the canvas containing
    // the visualization (size of the <div> element).
    var width = 1000,
        height = 600;

    // We create a color scale.
    var color = d3.scale.category10();

    // We create a force-directed dynamic graph layout.
    var force = d3.layout.force()
        .charge(-200)
        .linkDistance(60)
        .size([width, height]);

    // In the <div> element, we create a <svg> graphic
    // that will contain our interactive visualization.
    var svg = d3.select("#d3-example").select("svg")
    if (svg.empty()) {
        svg = d3.select("#d3-example").append("svg")
                    .attr("width", width)
                    .attr("height", height);
    }
        
    // We load the JSON file.
    d3.json("explainGraph.json", function(error, graph) {
        // In this block, the file has been loaded
        // and the 'graph' object contains our graph.
        
        // We load the nodes and links in the force-directed
        // graph.
        force.nodes(graph.nodes)
            .links(graph.links)
            .start();

        // We create a <line> SVG element for each link
        // in the graph.
        var link = svg.selectAll(".link")
            .data(graph.links)
            .enter().append("line")
            .attr("class", "link");

        // We create a <circle> SVG element for each node
        // in the graph, and we specify a few attributes.
        var node = svg.selectAll(".node")
            .data(graph.nodes)
            .enter().append("circle")
            .attr("class", "node")
            .attr("r", 12)  // radius
            .style("fill", function(d) {
                // The node color depends on the club.
                return 'red'; //color('white'); //d.club); 
            })
            .call(force.drag);

        // The name of each node is the node number.
        node.append("text")
            .attr("dy", ".35em")
            .text(function(d) { return d.label; });

        // We bind the positions of the SVG elements
        // to the positions of the dynamic force-directed graph,
        // at each time step.
        force.on("tick", function() {
            link.attr("x1", function(d) { return d.source.x; })
                .attr("y1", function(d) { return d.source.y; })
                .attr("x2", function(d) { return d.target.x; })
                .attr("y2", function(d) { return d.target.y; });

            node.attr("cx", function(d) { return d.x; })
                .attr("cy", function(d) { return d.y; });
        });
    });
});
```
